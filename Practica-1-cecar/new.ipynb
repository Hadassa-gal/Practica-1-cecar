```python
import yfinance as yf
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
#fecha de hoy
fecha_hoy = datetime.today().strftime("%Y-%m-%d")
```

```python
print(fecha_hoy)
```

```python
#decargar el precio del dolar desde el 2020 hasta hoy
dolar=yf.download("COP=X", start="2020-01-01", end=fecha_hoy)
```

```python
dolar
```

```python
dolar_1 = dolar[[('Open', 'COP=X'), ('Close', 'COP=X')]]
```

```python
dolar_1
```

```python
dolar_1["promedio"]=(dolar_1[('Open', 'COP=X')] + dolar_1[('Close', 'COP=X')])/2
```

```python
dolar_1.tail()
```

```python
dolar_mensual=dolar_1.resample('ME').mean()
```

```python
dolar_mensual
```

```python
#crear grafica de la evolución del precio del dolar
plt.figure(figsize=(12,6))
plt.plot(dolar_mensual.index,dolar_mensual,marker='o')
plt.grid(True)
plt.show()
```
